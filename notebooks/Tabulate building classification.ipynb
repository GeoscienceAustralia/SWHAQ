We need to evaluate the proportion that would be eligible for retrofit programs. For this, we use the lower 5 deciles of the ADRI index. ADRI is provided at SA2 level, so we need to first load the SA2 level ADRI data and assign eligible/not eligible to each asset based on it's SA2 code. 

```python
ADRIdf = pd.read_csv(r"X:\georisk\HaRIA_B_Wind\projects\qfes_swha\data\DRAFT DATA STRUCTURES\1. Work Unit Assessment\SOUTH EAST QUEENSLAND\ADRI_targeted_retrofit_eligability.csv")
eligible = list(ADRIdf[ADRIdf['Targeted_Retrofit']=='Eligible']['SA2_MAIN16'].astype('int64'))

df2 = df.loc[df['SA2_CODE'].isin(eligible)]
```

```python
table2 = df2.pivot_table(index=['LGA_NAME', 'AS4055_CLASS'],
                columns='WIND_VULNERABILITY_FUNCTION_ID',
                   aggfunc='size',
                   fill_value=0,)
```

```python
(table2/table).fillna(0).to_excel(r"X:\georisk\HaRIA_B_Wind\projects\qfes_swha\data\derived\exposure\2021\AS4055_counts.retrofit_eligible.xlsx")
```

```python
df['Eligibility'] = "Ineligible"
df.loc[df.SA2_CODE.isin(eligible),'Eligibility'] = "Eligible"
```

```python
table3 = df.pivot_table(index=['LGA_NAME', 'AS4055_CLASS'], columns=['WIND_VULNERABILITY_FUNCTION_ID', 'Eligibility'], aggfunc='size', fill_value=0).to_excel(r"X:\georisk\HaRIA_B_Wind\projects\qfes_swha\data\derived\exposure\2021\AS4055_counts.retrofit_eligible.v2.xlsx")
```
