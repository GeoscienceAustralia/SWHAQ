```python
dg = pydot.Dot(graph_name="TC wind hazard projection workflow", graph_type='digraph', rankdir="UD",)


def make_source_node(g, name, source, url):
    n = pydot.Node(name)
    n.set_shape('invhouse')
    n.set_label(f"{name}\n{source}")
    n.set_style('filled')
    n.set_fillcolor('green')
    n.set_URL(url)
    g.add_node(n)
    return n

def make_delivery_node(g, name, service):
    n = pydot.Node(name)
    n.set_shape("invhouse")
    n.set_label(f"{name}\n{service}")
    n.set_style('filled')
    n.set_fillcolor('red')
    g.add_node(n)
    return n

def make_node(g, name, shape):
    cur_node = pydot.Node(name)
    cur_node.set_shape(shape)
    g.add_node(cur_node)
    return cur_node

def make_link(g, a_node, b_node, label = None, width = 1, style='solid'):
    cur_edge = pydot.Edge(a_node,b_node)
    cur_edge.set_penwidth(width)
    cur_edge.set_style(style)
    if label is not None: cur_edge.set_label(label)
    g.add_edge(cur_edge)
    return cur_edge

def make_data_node(g, name, shape='folder', fillcolor='lightgreen'):
    cur_node = pydot.Node(name)
    cur_node.set_shape(shape)
    cur_node.set_style('filled')
    cur_node.set_fillcolor(fillcolor)
    g.add_node(cur_node)
    
    return cur_node

def attribute_node(g, name):
    n = pydot.Node(name)
    n.set_shape('Mrecord')
    g.add_node(n)
    return n

def collection_node(g, name, atts):
    n = pydot.Node(name)
    n.set_style("Mrecord")
    n.set_style('filled')
    n.set_fillcolor("lightgrey")
    rowfmt = "{{{{{" + "}}}|{{{".join([':s']*len(l)) + "}}}}}"
    n.set_label(rowfmt.format(*l))
    g.add_node(n)
    return n

def make_process_node(g, name, desc, fillcolor='lightblue'):
    cur_node = pydot.Node(desc)
    cur_node.set_shape('cds')
    cur_node.set_style('filled')
    cur_node.set_label(name)
    cur_node.set_fillcolor(fillcolor)
    g.add_node(cur_node)
    return cur_node
```
