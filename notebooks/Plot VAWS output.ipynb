```python
import os
import sys
import numpy as np
import seaborn as sns
import h5py

import matplotlib.pyplot as plt
sns.set_context("talk")
```

```python
filename = r"X:\georisk\HaRIA_B_Wind\projects\qfes_swha\data\derived\vulnerability\results.h5"
fh = h5py.File(filename, 'r')
```

```python
ws = fh['wind_speeds'][:]
meandi = fh['vulnerability']['mean_di'][:]
di = fh['house']['di'][:]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.scatter(np.tile(ws, (100, 1)), di.T, marker='.', c='k', s=5)
ax.scatter(ws, meandi, c='r', s=5, label="Mean")
ax.legend(loc=0)
ax.set_ylim((0, 1))
ax.set_ylabel("Damage Index")
ax.set_xlabel("Impact wind speed [m/s]")
```
