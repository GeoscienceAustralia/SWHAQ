List of stations to use in the extraction. These are Amberley, Archerfield, Brisbane Regional Office, Brisbane Airport, Coolangatta, Gold Coast Seaway, Sunshine Coast Airport, Tewantin, Redcliffe and Nambour. I have excluded Cape Moreton and Double Island Point from this analysis due to the known local effects that result in high wind gusts at those sites (they sit well above surrounding water on headlands, so are not representative of 10-m AGL wind speeds).

```python
stnList = [39083, 39085, 39112, 39123, 39128, 40040, 40043, 40068, 40004, 40211, 40214, 40223, 40717, 40764, 40861, 40908, 40958, 40988,]
```

Load the previously extracted data, and filter to the set of stations we want to examine

```python
df = pd.read_csv("X:/georisk/HaRIA_B_Wind/data/derived/tcobs/stncpa_obs.csv")
df = df[df.stnNum.isin(stnList)]
```

Group by the TC ID number, find the index of the maximum gust wind speed values and sort that set of values by the date/time of the TC event

```python
summary = df.loc[df.groupby("TCIDnum")['gust'].idxmax().values].sort_values('dtTC')
```

```python
summary.sort_values('gust')
```

Save the summary table to a csv for further use

```python
summary.to_csv("X:/georisk/HaRIA_B_Wind/data/derived/tcobs/maxwind_TC.allsites.csv", index=False)
```
